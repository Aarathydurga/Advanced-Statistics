The breaking strength of gunny bags used for packaging cement is normally distributed with a mean of 5 kg per sq. centimeter and a standard deviation of 1.5 kg per sq. centimeter. The quality team of the cement company wants to know the following about the packaging material to better understand wastage or pilferage within the supply chain; Answer the questions below based on the given information; (Provide an appropriate visual representation of your answers, without which marks will be deducted)

```python


# Parameters of the normal distribution
mean = 5  # Mean value
std_dev = 1.5  # Standard deviation

# Generate data points for x-axis
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 100)

# Calculate corresponding y-values using probability density function (PDF)
y = norm.pdf(x, loc=mean, scale=std_dev)

# Plotting the normal distribution graph
plt.plot(x, y)
plt.xlabel('Breaking Strength (kg per sq. cm)')
plt.ylabel('Probability Density')
plt.title('Normal Distribution of Breaking Strength')
plt.grid(True)
plt.show()
```

 What proportion of the gunny bags have a breaking strength less than 3.17 kg per sq cm?

```python
stats.norm.cdf(3.17 ,loc=5,scale=1.5)
```

```python
# Generate data points for x-axis
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 100)

# Calculate corresponding y-values using probability density function (PDF)
y = norm.pdf(x, loc=mean, scale=std_dev)
# Calculate the proportion of gunny bags with breaking strength less than 3.17 kg per sq cm
proportion = norm.cdf(3.17, loc=mean, scale=std_dev)

# Plotting the normal distribution graph
plt.plot(x, y)
plt.xlabel('Breaking Strength (kg per sq. cm)')
plt.ylabel('Probability Density')
plt.title('Normal Distribution of Breaking Strength')

# Plotting the shaded area representing the proportion
plt.fill_between(x, y, where=(x < 3.17), color='green', alpha=0.5)

# Display the proportion as text on the graph
plt.text(3.17, 0.02, f'Proportion: {proportion:.2f}', ha='center', va='bottom', color='red')

plt.grid(True)
plt.show()
```

Problem 4:

 Grades of the final examination in a training course are found to be normally distributed, with a mean of 77 and a standard deviation of 8.5. Based on the given information answer the questions below.


4.1 What is the probability that a randomly chosen student gets a grade below 85 on this exam?

```python
# Parameters of the normal distribution
mean = 77  # Mean value
std_dev = 8.5  # Standard deviation

# Calculate the probability that a student gets a grade below 85
probability_below_85 = norm.cdf(85, loc=mean, scale=std_dev)

print(f"The probability of getting a grade below 85 is: {probability_below_85:.4f}")
```

4.2 What is the probability that a randomly selected student scores between 65 and 87?

```python
# Calculate the probability that a student scores between 65 and 87
probability_between = norm.cdf(87, loc=mean, scale=std_dev) - norm.cdf(65, loc=mean, scale=std_dev)

print(f"The probability of scoring between 65 and 87 is: {probability_between:.4f}")
```

4.3 What should be the passing cut-off so that 75% of the students clear the exam?

```python
# Calculate the passing cut-off for 75% of the students
passing_cutoff = norm.ppf(0.75, loc=mean, scale=std_dev)

print(f"The passing cut-off for 75% of the students is: {passing_cutoff:.2f}")
```

Zingaro stone printing is a company that specializes in printing images or patterns on polished or unpolished stones. However, for the optimum level of printing of the image the stone surface has to have a Brinell's hardness index of at least 150. Recently, Zingaro has received a batch of polished and unpolished stones from its clients. Use the data provided to answer the following (assuming a 5% significance level);

 

5.1 Earlier experience of Zingaro with this particular client is favorable as the stone surface was found to be of adequate hardness. However, Zingaro has reason to believe now that the unpolished stones may not be suitable for printing. Do you think Zingaro is justified in thinking so?

```python
import pandas as pd
from scipy import stats

# Read the dataset from CSV
data = pd.read_csv("Zingaro_Company.csv")
```

We can calculate the sample mean and sample standard deviation of the unpolished stones dataset and perform a one-sample t-test. Using a significance level of 0.05, we can compare the calculated t_statistic and p_value to determine if there is enough evidence to reject the null hypothesis.

The null hypothesis (H0) would be that the mean hardness of the unpolished stones is at least 150.
The Alternative hypothesis (H1) would be that the mean hardness is less than 150.

```python
unpolished_stones = data["Unpolished "]
polished_stones = data["Treated and Polished"]

# 5.1 Hypothesis test for unpolished stones
mean_unpolished = unpolished_stones.mean()
t_statistic, p_value = stats.ttest_1samp(unpolished_stones, 150)

# Print the results
print("T-Statistic:", t_statistic)
print("P-Value:", p_value)


if p_value < 0.05:
    print("Zingaro is justified in thinking that unpolished stones may not be suitable for printing.")
    print("Reject the null hypothesis")
else:
    print("Zingaro's belief about unpolished stones may not be justified.")
    print("Fail to reject the null hypothesis")
```

5.2 Is the mean hardness of the polished and unpolished stones the same?

 To determine if the mean hardness of the polished and unpolished stones is the same, we can conduct a hypothesis test. The null hypothesis (H0) would be that the mean hardness of the polished and unpolished stones is equal, and the alternative hypothesis (H1) would be that the mean hardness is different.

```python
# 5.2 Hypothesis test for mean hardness
t_statistic, p_value = stats.ttest_ind(unpolished_stones, polished_stones)

# Print the results
print("T-Statistic:", t_statistic)
print("P-Value:", p_value/2)


if p_value < 0.05:
    print("The mean hardness of polished and unpolished stones are different.")
    print("Reject the null hypothesis")
else:
    print("The mean hardness of polished and unpolished stones are the same.")
    print("Fail to reject the null hypothesis")
```

# Aquarius health club, one of the largest and most popular cross-fit gyms in the country has been advertising a rigorous program for body conditioning. The program is considered successful if the candidate is able to do more than 5 push-ups, as compared to when he/she enrolled in the program. Using the sample data provided can you conclude whether the program is successful? (Consider the level of Significance as 5%)

Note that this is a problem of the paired-t-test. Since the claim is that the training will make a difference of more than 5, the null and alternative hypotheses must be formed accordingly.

 

```python
import pandas as pd
import scipy.stats as stats

# Read the dataset from CSV
data = pd.read_csv('Aquarius_gym.csv')
before = data['Before']
after = data['After']

# Perform paired t-test
t_statistic, p_value = stats.ttest_rel(after, before)

# Set the significance level
alpha = 0.05

# Null hypothesis: There is no significant difference in the number of push-ups before and after the program.
# Alternative hypothesis: The program is successful if the candidates can do more than 5 push-ups compared to when they enrolled.

# Print the t-statistic and p-value
print("T-Statistic:", t_statistic)
print("P-Value:", p_value)

# Compare the p-value with the significance level
if p_value < alpha:
    print("Reject the null hypothesis. The program is considered successful.")
else:
    print("Fail to reject the null hypothesis. The program is not considered successful.")
    print("There is no significant difference in the number of push-ups before and after the program.")
```
